```python
from itertools import combinations
from geopy.distance import geodesic
```

```python
kota = [
    "Kota Sukabumi",
    "Kota Bogor",
    "Kota Bandung",
    "Kota Cirebon",
    "Kota Bekasi",
    "Kota Depok",
    "Kota Cimahi",
    "Kota Tasikmalaya",
    "Kota Banjar",
]
```

```python
for k in kota:
    sisa_kota = [i for i in kota if i != k]
    kombinasi = list(combinations(sisa_kota, 8))
    print(f"Kombinasi untuk {k}:")
    for pasangan in kombinasi:
        output = '\n'.join('-'.join((k, p)) for p in pasangan)
        print(output)
    print("\n")
```

```python
def kota_bandung():
    kota_coords = {
        "Kota Sukabumi": (-6.928321402587708, 106.93293025148348),
        "Kota Bogor": (-6.596566663876446, 106.8037668905907),
        "Kota Bandung": (-6.911728885675809, 107.61798090819836),
        "Kota Cirebon": (-6.7358771403869575, 108.54273783107872),
        "Kota Bekasi": (-6.237608700273207, 106.97188794581838),
        "Kota Depok": (-6.404506760616969, 106.79697234762511),
        "Kota Cimahi": (-6.882252265828945, 107.54206579280836),
        "Kota Tasikmalaya": (-7.354854747830193, 108.21698654797164),
        "Kota Banjar": (-7.366750116329321, 108.53783174690206),
    }
    return kota_coords
kota_coords = kota_bandung()
```

```python
def hitung_jarak(kota1, kota2, kota_coords):
    return geodesic(kota_coords[kota1], kota_coords[kota2]).kilometers
```

```python
no = 0
for k in kota:
    no = no + 1
    sisa_kota = [i for i in kota if i != k]
    kombinasi = list(combinations(sisa_kota, 8))
    print(f"Kombinasi untuk {k} {no}:")
    for pasangan in kombinasi:
        output = '\n'.join(f"{k}-{p} = {int(hitung_jarak(k, p, kota_coords))} km" for p in pasangan)
        print(output)
    print("\n")
```
